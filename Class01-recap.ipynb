## 3) Type casting and User Input

- **Type casting** = การแปลงประเภทของตัวแปล จากประเภทหนึ่งไปอีกเป็นประเภทหนึ่ง เช่น ```int``` >>> ```str``` หรือ ```float``` >>> ```int()```

### 3.1) **int** >>> **str**

````python
num = 10
print(type(num)) # num = 10
````

````python
num_change = str(num)
print(type(num_change)) # num = "10"
````

### 3.2) **float** >>> **int**

````python
num2 = 12.5251

print("Before:", num2)
print(type(num2))
````

````python
num2_change = int(num2)

print("After:", num2_change)
print(type(num2))
````

### 3.3) **float** >>> **int**

````python
num3 = 120
print("Before:", num3)
print(type(num3))
````

````python
num3_change = float(num3)
print("After:", num3_change)
print(type(num3_change))
````

## 5) If-else statement

**Normal If-else**

````python
gender = "woman" 

if gender == "man":
    print("This is a man.")
else:
    print("I don't konw.")
````

**If-elif-else**

````python
num1 = 100

if num1 > 100:
    print("Greater than 100")
elif 100 >= num1 > 50:
    print("Still greater than 50 or maybe 100")
else:
    print("Lower than half")
````

**Shortcut If-else**

````python
age = 21

result = "adult" if age > 20 else "kid"
print(result)
````

**If only**

````python
condition = True

if condition:
    print("In 'if' condition")

print("End program")
````

**Switch case**

````python
num2 = 1

match num2:
    case 1:
        print("one")
    case 2:
        print("two")
    case 3:
        print("three")
    case 4:
        print("four")
    case 5:
        print("five")
````

````python
fruit = "banana"

match fruit:
    case "orange":
        print("one")
    case "melon":
        print("two")
    case "lemon":
        print("three")
    case "grapes":
        print("four")
    case "banana":
        print("five")
````
